```python
import pandas as pd
from pathlib import Path


def load_csv_columns_with_sample(data_dir: str = "./data") -> dict:
    """
    data Ìè¥ÎçîÏóê ÏûàÎäî Î™®Îì† CSV ÌååÏùºÏùÑ Î∂àÎü¨ÏôÄÏÑú Ïª¨ÎüºÍ≥º ÏÉòÌîå Îç∞Ïù¥ÌÑ∞(1Ï§Ñ)Î•º Ï∂úÎ†•ÌïòÎäî Ìï®Ïàò

    Args:
        data_dir: CSV ÌååÏùºÎì§Ïù¥ ÏûàÎäî ÎîîÎ†âÌÜ†Î¶¨ Í≤ΩÎ°ú (Í∏∞Î≥∏Í∞í: "./data")

    Returns:
        dict: ÌååÏùºÎ™ÖÏùÑ keyÎ°ú, Ïª¨Îüº Î¶¨Ïä§Ìä∏ÏôÄ ÏÉòÌîå Îç∞Ïù¥ÌÑ∞Î•º valueÎ°ú ÌïòÎäî ÎîïÏÖîÎÑàÎ¶¨
    """
    data_path = Path(data_dir)
    csv_files = list(data_path.glob("*.csv"))

    if not csv_files:
        print(f"'{data_dir}' Ìè¥ÎçîÏóê CSV ÌååÏùºÏù¥ ÏóÜÏäµÎãàÎã§.")
        return {}

    columns_dict = {}

    for csv_file in csv_files:
        try:
            # Ïó¨Îü¨ Ïù∏ÏΩîÎî© ÏãúÎèÑ
            encodings = ['utf-8-sig', 'utf-8', 'cp949', 'euc-kr']
            df = None
            
            for encoding in encodings:
                try:
                    df = pd.read_csv(csv_file, nrows=1, encoding=encoding)
                    break
                except UnicodeDecodeError:
                    continue
            
            if df is None:
                print(f"'{csv_file.name}' ÌååÏùº ÏùΩÍ∏∞ Ïò§Î•ò: Ïù∏ÏΩîÎî©ÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§.")
                continue
            
            columns_dict[csv_file.name] = {
                'columns': df.columns.tolist(),
                'sample': df.iloc[0].to_dict() if len(df) > 0 else {}
            }

            print(f"\n{'='*80}")
            print(f"üìÑ ÌååÏùºÎ™Ö: {csv_file.name}")
            print(f"üìä Ïª¨Îüº Í∞úÏàò: {len(df.columns)}")
            print(f"üìã Ïª¨Îüº Î™©Î°ù: {df.columns.tolist()}")
            print(f"\nüîç ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ (1Ìñâ):")
            for col, value in df.iloc[0].items():
                print(f"  - {col}: {value}")
            print("="*80)
            
        except Exception as e:
            print(f"'{csv_file.name}' ÌååÏùº ÏùΩÍ∏∞ Ïò§Î•ò: {e}")

    return columns_dict



if __name__ == "__main__":
    # ÏÇ¨Ïö© ÏòàÏãú
    columns = load_csv_columns_with_sample("/mnt/c/Users/rladl/Desktop/bigcontest_2025/2025_shcard_bigcontest/data")
```
