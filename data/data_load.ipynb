```python
import pandas as pd
from pathlib import Path


def load_csv_columns_with_sample(data_dir: str = "./data") -> dict:
    """
    data 폴더에 있는 모든 CSV 파일을 불러와서 컬럼과 샘플 데이터(1줄)를 출력하는 함수

    Args:
        data_dir: CSV 파일들이 있는 디렉토리 경로 (기본값: "./data")

    Returns:
        dict: 파일명을 key로, 컬럼 리스트와 샘플 데이터를 value로 하는 딕셔너리
    """
    data_path = Path(data_dir)
    csv_files = list(data_path.glob("*.csv"))

    if not csv_files:
        print(f"'{data_dir}' 폴더에 CSV 파일이 없습니다.")
        return {}

    columns_dict = {}

    for csv_file in csv_files:
        try:
            # 여러 인코딩 시도
            encodings = ['utf-8-sig', 'utf-8', 'cp949', 'euc-kr']
            df = None
            
            for encoding in encodings:
                try:
                    df = pd.read_csv(csv_file, nrows=1, encoding=encoding)
                    break
                except UnicodeDecodeError:
                    continue
            
            if df is None:
                print(f"'{csv_file.name}' 파일 읽기 오류: 인코딩을 찾을 수 없습니다.")
                continue
            
            columns_dict[csv_file.name] = {
                'columns': df.columns.tolist(),
                'sample': df.iloc[0].to_dict() if len(df) > 0 else {}
            }

            print(f"\n{'='*80}")
            print(f"📄 파일명: {csv_file.name}")
            print(f"📊 컬럼 개수: {len(df.columns)}")
            print(f"📋 컬럼 목록: {df.columns.tolist()}")
            print(f"\n🔍 샘플 데이터 (1행):")
            for col, value in df.iloc[0].items():
                print(f"  - {col}: {value}")
            print("="*80)
            
        except Exception as e:
            print(f"'{csv_file.name}' 파일 읽기 오류: {e}")

    return columns_dict



if __name__ == "__main__":
    # 사용 예시
    columns = load_csv_columns_with_sample("/mnt/c/Users/rladl/Desktop/bigcontest_2025/2025_shcard_bigcontest/data")
```
