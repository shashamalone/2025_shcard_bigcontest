```python
import requests, math, random
from IPython.display import HTML, display

API_KEY = "NiCSGOCv9sUFIyekjbTsVrp22ZDmTvTDHaFuAVUpsP3ENj6wWcHvIfP3" 
QUERIES = [
    "coffee shop interior",
    "cafe exterior",
    "bakery storefront",
    "restaurant interior",
    "dessert shop",
    "bar counter",
    "korea",
]

def fetch_photos_multi(queries, total=6, page=1, orientation="portrait", color="brown"):
    """
    여러 쿼리를 순회하며 사진을 모아 총 total장 반환(중복 id 제거).
    한 쿼리당 균등하게 요청해서 골고루 섞이게 합니다.
    """
    per_query = math.ceil(total / len(queries))
    collected = []
    seen_ids = set()

    for q in queries:
        url = (
            "https://api.pexels.com/v1/search"
            f"?query={q}"
            f"&per_page={per_query}"
            f"&page={page}"
            f"&orientation={orientation}"
            f"&color={color}"
        )
        r = requests.get(url, headers={"Authorization": API_KEY})
        r.raise_for_status()
        photos = r.json().get("photos", [])

        for p in photos:
            if p["id"] in seen_ids:
                continue
            # 원본 객체에 이 사진이 어떤 쿼리로 들어왔는지 표시(렌더에서 사용)
            p["_topic"] = q
            collected.append(p)
            seen_ids.add(p["id"])
            if len(collected) >= total:
                break
        if len(collected) >= total:
            break

    # 수집 개수가 모자라면(검색 결과가 적은 쿼리 등) 그냥 모은 만큼 반환
    return collected[:total]

```
