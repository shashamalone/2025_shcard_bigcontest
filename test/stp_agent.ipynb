```python
import operator
from typing import List, TypedDict
from typing_extensions import Annotated
from langchain_core.messages import BaseMessage, HumanMessage
from langgraph.graph import StateGraph, END

# --- 1. 최상위 그래프의 상태 정의 ---
# 이 State 객체는 그래프의 모든 노드를 거치며 데이터와 작업 흐름을 관리합니다.

class SuperGraphState(TypedDict):
    """
    전체 워크플로우의 상태를 관리합니다.

    Attributes:
        messages: 전체 대화 및 작업 결과가 누적되는 리스트
        intent: 사용자의 질문 의도를 저장 ('comprehensive_strategy', 'content_guide', 'situational_tactic')
    """
    messages: Annotated[List[BaseMessage], operator.add]
    intent: str

# --- 2. 각 에이전트 팀의 역할 시뮬레이션 (플레이스홀더 함수) ---
# 실제로는 각 팀이 복잡한 서브그래프로 구성되지만, 여기서는 단일 함수로 역할을 대체합니다.

def team_1_analysis_placeholder(state: SuperGraphState) -> dict:
    """Team 1: 데이터 분석을 시뮬레이션합니다."""
    print("✅ [Team 1] 분석 시작...")
    analysis_result = "분석 결과: 이 가게의 핵심 고객은 20대 여성으로, 인스타그램 사용률이 매우 높음."
    return {"messages": [HumanMessage(content=analysis_result, name="Team_1_Analysis")]}

def team_3_execution_placeholder(state: SuperGraphState) -> dict:
    """Team 3: 실행 계획 및 콘텐츠 생성을 시뮬레이션합니다."""
    print("✅ [Team 3] 실행 계획 및 콘텐츠 생성 시작...")
    # Team 1의 분석 결과를 입력으로 받습니다.
    analysis_data = state['messages'][-1].content
    plan = (
        f"'{analysis_data}'에 기반하여,\n"
        "1. 인스타그램 릴스 챌린지 기획\n"
        "2. 해시태그: #성수동카페 #주말데이트\n"
        "3. 무드보드: 파스텔 톤의 감성적인 이미지 컨셉"
    )
    return {"messages": [HumanMessage(content=plan, name="Team_3_Execution")]}

def team_4_tactical_placeholder(state: SuperGraphState) -> dict:
    """Team 4: 상황 기반 전술 제안을 시뮬레이션합니다."""
    print("✅ [Team 4] 상황 분석 및 전술 제안 시작...")
    tactic = "긴급 전술 제안: 오늘 성동구에 비 예보가 있으니, '비 오는 날 1+1 음료' 이벤트를 즉시 인스타그램 스토리에 공지하세요."
    return {"messages": [HumanMessage(content=tactic, name="Team_4_Tactical")]}

# --- 3. 의도 분류기 및 라우터 정의 ---

def classify_intent(state: SuperGraphState) -> dict:
    """사용자의 첫 질문을 분석하여 의도를 분류합니다."""
    print("✅ [Gatekeeper] 사용자 의도 분류 시작...")
    user_query = state['messages'][-1].content.lower() # 소문자로 변환하여 비교

    # 실제로는 LLM을 사용하지만, 여기서는 키워드 기반으로 시뮬레이션합니다.
    if "비오는 날" in user_query or "행사" in user_query:
        intent = "situational_tactic"
    elif "인스타그램" in user_query or "가이드" in user_query or "홍보" in user_query:
        intent = "content_guide"
    else:
        intent = "comprehensive_strategy"

    print(f"🔀 분류된 의도: {intent}")
    return {"intent": intent}

def orchestrator(state: SuperGraphState) -> str:
    """분류된 의도에 따라 다음 작업 팀으로 라우팅합니다."""
    print("🔀 [Orchestrator] 다음 경로 결정...")
    intent = state.get("intent")
    if intent == "situational_tactic":
        # 상황 전술 제안은 Team 4로 직접 연결
        return "Team_4_Tactical"
    elif intent in ["comprehensive_strategy", "content_guide"]:
        # 종합 전략 및 콘텐츠 가이드는 먼저 Team 1의 분석을 거침
        return "Team_1_Analysis"
    else:
        # 예외 처리
        return "end"

# --- 4. 전체 그래프 구성 및 컴파일 ---

workflow = StateGraph(SuperGraphState)

# 노드(작업 단위) 추가
workflow.add_node("Intent_Classifier", classify_intent)
workflow.add_node("Team_1_Analysis", team_1_analysis_placeholder)
workflow.add_node("Team_3_Execution", team_3_execution_placeholder)
workflow.add_node("Team_4_Tactical", team_4_tactical_placeholder)

# 진입점(Entry Point) 설정
workflow.set_entry_point("Intent_Classifier")

# 조건부 엣지(라우팅) 설정
workflow.add_conditional_edges(
    "Intent_Classifier",
    orchestrator,
    {
        "Team_1_Analysis": "Team_1_Analysis",
        "Team_4_Tactical": "Team_4_Tactical",
        "end": END
    }
)

# 일반 엣지(작업 흐름) 설정
workflow.add_edge("Team_1_Analysis", "Team_3_Execution") # 분석 후에는 항상 실행 계획으로
workflow.add_edge("Team_3_Execution", END) # 실행 계획 후 종료
workflow.add_edge("Team_4_Tactical", END)   # 전술 제안 후 종료

# 그래프 컴파일
app = workflow.compile()

# --- 5. 테스트 실행 ---
def run_test_query(query: str):
    print(f"\n--- 🚀 테스트 시작: '{query}' ---")
    inputs = {"messages": [HumanMessage(content=query)]}
    final_state = app.invoke(inputs)
    final_message = final_state['messages'][-1].content
    print("\n--- 🏁 최종 결과물 ---")
    print(final_message)
    print("---------------------\n")

# 테스트 질문 실행
run_test_query("카페업종 가맹점 A - 카페 주요 방문 고객 특성에 따른 마케팅 채널 추천 및 홍보안 필요")
run_test_query("치킨업종 가맹점 E - 우리 치킨집 인스타그램 홍보 방안 가이드 좀 알려줘")
run_test_query("카페업종 가맹점 D - 비오는날의 전략을 고안해줘")

app
```
