```python
import operator
from typing import List, TypedDict
from typing_extensions import Annotated
from langchain_core.messages import BaseMessage, HumanMessage
from langgraph.graph import StateGraph, END

# --- 1. ÏµúÏÉÅÏúÑ Í∑∏ÎûòÌîÑÏùò ÏÉÅÌÉú Ï†ïÏùò ---
# Ïù¥ State Í∞ùÏ≤¥Îäî Í∑∏ÎûòÌîÑÏùò Î™®Îì† ÎÖ∏ÎìúÎ•º Í±∞ÏπòÎ©∞ Îç∞Ïù¥ÌÑ∞ÏôÄ ÏûëÏóÖ ÌùêÎ¶ÑÏùÑ Í¥ÄÎ¶¨Ìï©ÎãàÎã§.

class SuperGraphState(TypedDict):
    """
    Ï†ÑÏ≤¥ ÏõåÌÅ¨ÌîåÎ°úÏö∞Ïùò ÏÉÅÌÉúÎ•º Í¥ÄÎ¶¨Ìï©ÎãàÎã§.

    Attributes:
        messages: Ï†ÑÏ≤¥ ÎåÄÌôî Î∞è ÏûëÏóÖ Í≤∞Í≥ºÍ∞Ä ÎàÑÏ†ÅÎêòÎäî Î¶¨Ïä§Ìä∏
        intent: ÏÇ¨Ïö©ÏûêÏùò ÏßàÎ¨∏ ÏùòÎèÑÎ•º Ï†ÄÏû• ('comprehensive_strategy', 'content_guide', 'situational_tactic')
    """
    messages: Annotated[List[BaseMessage], operator.add]
    intent: str

# --- 2. Í∞Å ÏóêÏù¥Ï†ÑÌä∏ ÌåÄÏùò Ïó≠Ìï† ÏãúÎÆ¨Î†àÏù¥ÏÖò (ÌîåÎ†àÏù¥Ïä§ÌôÄÎçî Ìï®Ïàò) ---
# Ïã§Ï†úÎ°úÎäî Í∞Å ÌåÄÏù¥ Î≥µÏû°Ìïú ÏÑúÎ∏åÍ∑∏ÎûòÌîÑÎ°ú Íµ¨ÏÑ±ÎêòÏßÄÎßå, Ïó¨Í∏∞ÏÑúÎäî Îã®Ïùº Ìï®ÏàòÎ°ú Ïó≠Ìï†ÏùÑ ÎåÄÏ≤¥Ìï©ÎãàÎã§.

def team_1_analysis_placeholder(state: SuperGraphState) -> dict:
    """Team 1: Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑùÏùÑ ÏãúÎÆ¨Î†àÏù¥ÏÖòÌï©ÎãàÎã§."""
    print("‚úÖ [Team 1] Î∂ÑÏÑù ÏãúÏûë...")
    analysis_result = "Î∂ÑÏÑù Í≤∞Í≥º: Ïù¥ Í∞ÄÍ≤åÏùò ÌïµÏã¨ Í≥†Í∞ùÏùÄ 20ÎåÄ Ïó¨ÏÑ±ÏúºÎ°ú, Ïù∏Ïä§ÌÉÄÍ∑∏Îû® ÏÇ¨Ïö©Î•†Ïù¥ Îß§Ïö∞ ÎÜíÏùå."
    return {"messages": [HumanMessage(content=analysis_result, name="Team_1_Analysis")]}

def team_3_execution_placeholder(state: SuperGraphState) -> dict:
    """Team 3: Ïã§Ìñâ Í≥ÑÌöç Î∞è ÏΩòÌÖêÏ∏† ÏÉùÏÑ±ÏùÑ ÏãúÎÆ¨Î†àÏù¥ÏÖòÌï©ÎãàÎã§."""
    print("‚úÖ [Team 3] Ïã§Ìñâ Í≥ÑÌöç Î∞è ÏΩòÌÖêÏ∏† ÏÉùÏÑ± ÏãúÏûë...")
    # Team 1Ïùò Î∂ÑÏÑù Í≤∞Í≥ºÎ•º ÏûÖÎ†•ÏúºÎ°ú Î∞õÏäµÎãàÎã§.
    analysis_data = state['messages'][-1].content
    plan = (
        f"'{analysis_data}'Ïóê Í∏∞Î∞òÌïòÏó¨,\n"
        "1. Ïù∏Ïä§ÌÉÄÍ∑∏Îû® Î¶¥Ïä§ Ï±åÎ¶∞ÏßÄ Í∏∞Ìöç\n"
        "2. Ìï¥ÏãúÌÉúÍ∑∏: #ÏÑ±ÏàòÎèôÏπ¥Ìéò #Ï£ºÎßêÎç∞Ïù¥Ìä∏\n"
        "3. Î¨¥ÎìúÎ≥¥Îìú: ÌååÏä§ÌÖî ÌÜ§Ïùò Í∞êÏÑ±Ï†ÅÏù∏ Ïù¥ÎØ∏ÏßÄ Ïª®ÏÖâ"
    )
    return {"messages": [HumanMessage(content=plan, name="Team_3_Execution")]}

def team_4_tactical_placeholder(state: SuperGraphState) -> dict:
    """Team 4: ÏÉÅÌô© Í∏∞Î∞ò Ï†ÑÏà† Ï†úÏïàÏùÑ ÏãúÎÆ¨Î†àÏù¥ÏÖòÌï©ÎãàÎã§."""
    print("‚úÖ [Team 4] ÏÉÅÌô© Î∂ÑÏÑù Î∞è Ï†ÑÏà† Ï†úÏïà ÏãúÏûë...")
    tactic = "Í∏¥Í∏â Ï†ÑÏà† Ï†úÏïà: Ïò§Îäò ÏÑ±ÎèôÍµ¨Ïóê ÎπÑ ÏòàÎ≥¥Í∞Ä ÏûàÏúºÎãà, 'ÎπÑ Ïò§Îäî ÎÇ† 1+1 ÏùåÎ£å' Ïù¥Î≤§Ìä∏Î•º Ï¶âÏãú Ïù∏Ïä§ÌÉÄÍ∑∏Îû® Ïä§ÌÜ†Î¶¨Ïóê Í≥µÏßÄÌïòÏÑ∏Ïöî."
    return {"messages": [HumanMessage(content=tactic, name="Team_4_Tactical")]}

# --- 3. ÏùòÎèÑ Î∂ÑÎ•òÍ∏∞ Î∞è ÎùºÏö∞ÌÑ∞ Ï†ïÏùò ---

def classify_intent(state: SuperGraphState) -> dict:
    """ÏÇ¨Ïö©ÏûêÏùò Ï≤´ ÏßàÎ¨∏ÏùÑ Î∂ÑÏÑùÌïòÏó¨ ÏùòÎèÑÎ•º Î∂ÑÎ•òÌï©ÎãàÎã§."""
    print("‚úÖ [Gatekeeper] ÏÇ¨Ïö©Ïûê ÏùòÎèÑ Î∂ÑÎ•ò ÏãúÏûë...")
    user_query = state['messages'][-1].content.lower() # ÏÜåÎ¨∏ÏûêÎ°ú Î≥ÄÌôòÌïòÏó¨ ÎπÑÍµê

    # Ïã§Ï†úÎ°úÎäî LLMÏùÑ ÏÇ¨Ïö©ÌïòÏßÄÎßå, Ïó¨Í∏∞ÏÑúÎäî ÌÇ§ÏõåÎìú Í∏∞Î∞òÏúºÎ°ú ÏãúÎÆ¨Î†àÏù¥ÏÖòÌï©ÎãàÎã§.
    if "ÎπÑÏò§Îäî ÎÇ†" in user_query or "ÌñâÏÇ¨" in user_query:
        intent = "situational_tactic"
    elif "Ïù∏Ïä§ÌÉÄÍ∑∏Îû®" in user_query or "Í∞ÄÏù¥Îìú" in user_query or "ÌôçÎ≥¥" in user_query:
        intent = "content_guide"
    else:
        intent = "comprehensive_strategy"

    print(f"üîÄ Î∂ÑÎ•òÎêú ÏùòÎèÑ: {intent}")
    return {"intent": intent}

def orchestrator(state: SuperGraphState) -> str:
    """Î∂ÑÎ•òÎêú ÏùòÎèÑÏóê Îî∞Îùº Îã§Ïùå ÏûëÏóÖ ÌåÄÏúºÎ°ú ÎùºÏö∞ÌåÖÌï©ÎãàÎã§."""
    print("üîÄ [Orchestrator] Îã§Ïùå Í≤ΩÎ°ú Í≤∞Ï†ï...")
    intent = state.get("intent")
    if intent == "situational_tactic":
        # ÏÉÅÌô© Ï†ÑÏà† Ï†úÏïàÏùÄ Team 4Î°ú ÏßÅÏ†ë Ïó∞Í≤∞
        return "Team_4_Tactical"
    elif intent in ["comprehensive_strategy", "content_guide"]:
        # Ï¢ÖÌï© Ï†ÑÎûµ Î∞è ÏΩòÌÖêÏ∏† Í∞ÄÏù¥ÎìúÎäî Î®ºÏ†Ä Team 1Ïùò Î∂ÑÏÑùÏùÑ Í±∞Ïπ®
        return "Team_1_Analysis"
    else:
        # ÏòàÏô∏ Ï≤òÎ¶¨
        return "end"

# --- 4. Ï†ÑÏ≤¥ Í∑∏ÎûòÌîÑ Íµ¨ÏÑ± Î∞è Ïª¥ÌååÏùº ---

workflow = StateGraph(SuperGraphState)

# ÎÖ∏Îìú(ÏûëÏóÖ Îã®ÏúÑ) Ï∂îÍ∞Ä
workflow.add_node("Intent_Classifier", classify_intent)
workflow.add_node("Team_1_Analysis", team_1_analysis_placeholder)
workflow.add_node("Team_3_Execution", team_3_execution_placeholder)
workflow.add_node("Team_4_Tactical", team_4_tactical_placeholder)

# ÏßÑÏûÖÏ†ê(Entry Point) ÏÑ§Ï†ï
workflow.set_entry_point("Intent_Classifier")

# Ï°∞Í±¥Î∂Ä Ïó£ÏßÄ(ÎùºÏö∞ÌåÖ) ÏÑ§Ï†ï
workflow.add_conditional_edges(
    "Intent_Classifier",
    orchestrator,
    {
        "Team_1_Analysis": "Team_1_Analysis",
        "Team_4_Tactical": "Team_4_Tactical",
        "end": END
    }
)

# ÏùºÎ∞ò Ïó£ÏßÄ(ÏûëÏóÖ ÌùêÎ¶Ñ) ÏÑ§Ï†ï
workflow.add_edge("Team_1_Analysis", "Team_3_Execution") # Î∂ÑÏÑù ÌõÑÏóêÎäî Ìï≠ÏÉÅ Ïã§Ìñâ Í≥ÑÌöçÏúºÎ°ú
workflow.add_edge("Team_3_Execution", END) # Ïã§Ìñâ Í≥ÑÌöç ÌõÑ Ï¢ÖÎ£å
workflow.add_edge("Team_4_Tactical", END)   # Ï†ÑÏà† Ï†úÏïà ÌõÑ Ï¢ÖÎ£å

# Í∑∏ÎûòÌîÑ Ïª¥ÌååÏùº
app = workflow.compile()

# --- 5. ÌÖåÏä§Ìä∏ Ïã§Ìñâ ---
def run_test_query(query: str):
    print(f"\n--- üöÄ ÌÖåÏä§Ìä∏ ÏãúÏûë: '{query}' ---")
    inputs = {"messages": [HumanMessage(content=query)]}
    final_state = app.invoke(inputs)
    final_message = final_state['messages'][-1].content
    print("\n--- üèÅ ÏµúÏ¢Ö Í≤∞Í≥ºÎ¨º ---")
    print(final_message)
    print("---------------------\n")

# ÌÖåÏä§Ìä∏ ÏßàÎ¨∏ Ïã§Ìñâ
run_test_query("Ïπ¥ÌéòÏóÖÏ¢Ö Í∞ÄÎßπÏ†ê A - Ïπ¥Ìéò Ï£ºÏöî Î∞©Î¨∏ Í≥†Í∞ù ÌäπÏÑ±Ïóê Îî∞Î•∏ ÎßàÏºÄÌåÖ Ï±ÑÎÑê Ï∂îÏ≤ú Î∞è ÌôçÎ≥¥Ïïà ÌïÑÏöî")
run_test_query("ÏπòÌÇ®ÏóÖÏ¢Ö Í∞ÄÎßπÏ†ê E - Ïö∞Î¶¨ ÏπòÌÇ®Ïßë Ïù∏Ïä§ÌÉÄÍ∑∏Îû® ÌôçÎ≥¥ Î∞©Ïïà Í∞ÄÏù¥Îìú Ï¢Ä ÏïåÎ†§Ï§ò")
run_test_query("Ïπ¥ÌéòÏóÖÏ¢Ö Í∞ÄÎßπÏ†ê D - ÎπÑÏò§ÎäîÎÇ†Ïùò Ï†ÑÎûµÏùÑ Í≥†ÏïàÌï¥Ï§ò")

app
```
